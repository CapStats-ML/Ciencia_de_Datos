```python
# ----------------- K-Fold Cross Validation -----------------
kf = KFold(n_splits=5, shuffle=True, random_state=1998)

# Inicializaci√≥n de variables para almacenar resultados
accuracies = []
roc_auc_scores = []

# Inicializar la figura para las curvas ROC
plt.figure(figsize=(12, 8))

# Definir una paleta de colores
colors = plt.cm.viridis(np.linspace(0, 1, kf.get_n_splits()))

# Realizar K-Fold Cross Validation
for fold, (train_index, test_index) in enumerate(kf.split(XFAMD)):
    # Dividir los datos en conjunto de entrenamiento y prueba
    X_train, X_test = XFAMD.iloc[train_index], XFAMD.iloc[test_index]
    y_train, y_test = YFAMD.iloc[train_index], YFAMD.iloc[test_index]

    # Crear el modelo GLM con familia binomial
    X_train_scaled_sm = sm.add_constant(X_train)
    X_test_scaled_sm = sm.add_constant(X_test)

    # Ajustar el modelo GLM
    glm_binom = sm.GLM(y_train, X_train_scaled_sm, family=sm.families.Binomial())
    res = glm_binom.fit()

    # Realizar predicciones sobre el conjunto de prueba
    y_pred_prob_glm = res.predict(X_test_scaled_sm)
    y_pred_glm = (y_pred_prob_glm > 0.5).astype(int)

    # Evaluaci√≥n: Precisi√≥n
    accuracy = metrics.accuracy_score(y_test, y_pred_glm)
    accuracies.append(accuracy)

    # Evaluaci√≥n: Curva ROC y AUC
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob_glm)
    roc_auc = auc(fpr, tpr)
    roc_auc_scores.append(roc_auc)

    # Graficar la curva ROC para cada pliegue
    plt.plot(fpr, tpr, color=colors[fold], lw=2, label=f'Fold {fold + 1} (AUC = {roc_auc:.2f})')

avg_accuracy = np.mean(accuracies)
avg_roc_auc = np.mean(roc_auc_scores)


# Graficar la l√≠nea diagonal (clasificador aleatorio)
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', lw=2, label="Clasificador aleatorio")

# Ajustes del gr√°fico
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title(f'Curvas ROC - Validaci√≥n Cruzada (AUC Promedio = {np.mean(roc_auc_scores):.2f})')
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


print(f"Precisi√≥n promedio: {avg_accuracy:.2f}")
print(f"AUC promedio : {avg_roc_auc:.2f}")
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.genmod.families import links
from sklearn.model_selection import KFold
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# Datos de ejemplo (XFAMD y YFAMD deben estar definidos)
# XFAMD = ...
# YFAMD = ...

# Definir las funciones de enlace como objetos
link_functions = {
    'logit': links.Logit(),
    'probit': links.Probit(),
    'cloglog': links.CLogLog()
}

# Valores de alpha para la regularizaci√≥n
alphas = [0.001, 0.01, 0.05]

# Configuraci√≥n de K-Fold Cross Validation
kf = KFold(n_splits=5, shuffle=True, random_state=1998)

# Almacenar resultados
results = []

# Configurar la grilla de gr√°ficos
fig, axes = plt.subplots(3, 3, figsize=(18, 18))
fig.suptitle('Curvas ROC por Funci√≥n de Enlace y Alpha', fontsize=16)

# Definir una paleta de colores para los pliegues
colors = plt.cm.viridis(np.linspace(0, 1, kf.get_n_splits()))

# Realizar la b√∫squeda de hiperpar√°metros
for i, (link_name, link) in enumerate(link_functions.items()):
    for j, alpha in enumerate(alphas):
        accuracies = []
        roc_auc_scores = []
        
        # Seleccionar el subgr√°fico actual
        ax = axes[i, j]
        
        for fold, (train_index, test_index) in enumerate(kf.split(XFAMD)):
            # Dividir los datos
            X_train, X_test = XFAMD.iloc[train_index], XFAMD.iloc[test_index]
            y_train, y_test = YFAMD.iloc[train_index], YFAMD.iloc[test_index]

            # A√±adir constante
            X_train_scaled_sm = sm.add_constant(X_train)
            X_test_scaled_sm = sm.add_constant(X_test)

            # Crear el modelo GLM con la funci√≥n de enlace actual
            glm_binom = sm.GLM(y_train, X_train_scaled_sm, 
                               family=sm.families.Binomial(link=link))

            # Ajustar el modelo con regularizaci√≥n
            res = glm_binom.fit_regularized(alpha=alpha)

            # Realizar predicciones
            y_pred_prob_glm = res.predict(X_test_scaled_sm)
            y_pred_glm = (y_pred_prob_glm > 0.5).astype(int)

            # Calcular m√©tricas
            accuracy = metrics.accuracy_score(y_test, y_pred_glm)
            fpr, tpr, _ = roc_curve(y_test, y_pred_prob_glm)
            roc_auc = auc(fpr, tpr)

            # Almacenar resultados del fold
            accuracies.append(accuracy)
            roc_auc_scores.append(roc_auc)

            # Graficar la curva ROC para cada pliegue
            ax.plot(fpr, tpr, color=colors[fold], lw=2, 
                    label=f'Fold {fold + 1} AUC = {roc_auc:.2f}')

        # Calcular promedios para el conjunto de hiperpar√°metros actual
        avg_accuracy = np.mean(accuracies)
        avg_roc_auc = np.mean(roc_auc_scores)

        # Guardar resultados
        results.append({
            'link': link_name,
            'alpha': alpha,
            'avg_accuracy': avg_accuracy,
            'avg_roc_auc': avg_roc_auc
        })

        # Graficar la l√≠nea diagonal (clasificador aleatorio)
        ax.plot([0, 1], [0, 1], linestyle='--', color='gray', lw=2, label="Clasificador aleatorio")

        # Ajustes del subgr√°fico
        ax.set_xlabel('Tasa de Falsos Positivos (FPR)')
        ax.set_ylabel('Tasa de Verdaderos Positivos (TPR)')
        ax.set_title(f'{link_name} (alpha={alpha})')
        ax.legend(loc='lower right')
        ax.grid(True, linestyle='--', alpha=0.6)

# Ajustar el layout de la grilla
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Convertir resultados a DataFrame
results_df = pd.DataFrame(results)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Datos de ejemplo (XDAT_encoded y YDAT deben estar definidos)
# XDAT_encoded = ...
# YDAT = ...

# Definir los valores de C (inverso de la fuerza de regularizaci√≥n) y los tipos de regularizaci√≥n
Cs = [0.01, 0.05, 0.1]  # Valores de C
penalties = ['l1', 'l2']  # Tipos de regularizaci√≥n

# Configuraci√≥n de K-Fold Cross Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Almacenar resultados
results = []

# Configurar la grilla de gr√°ficos
fig, axes = plt.subplots(len(penalties), len(Cs), figsize=(18, 12))
fig.suptitle('Curvas ROC por Tipo de Regularizaci√≥n y Valor de C', fontsize=16)

# Definir una paleta de colores para los pliegues
colors = plt.cm.viridis(np.linspace(0, 1, kf.get_n_splits()))

# Realizar la b√∫squeda de hiperpar√°metros
for i, penalty in enumerate(penalties):
    for j, C in enumerate(Cs):
        accuracies = []
        roc_auc_scores = []
        
        # Seleccionar el subgr√°fico actual
        ax = axes[i, j]
        
        for fold, (train_index, test_index) in enumerate(kf.split(XDAT_encoded)):
            # Dividir los datos usando .iloc para acceder a las filas por √≠ndices
            X_train, X_test = XDAT_encoded.iloc[train_index], XDAT_encoded.iloc[test_index]
            y_train, y_test = YDAT.iloc[train_index], YDAT.iloc[test_index]

            # Crear el modelo de regresi√≥n log√≠stica
            log_model = LogisticRegression(penalty=penalty, C=C, solver='liblinear')
            log_model.fit(X_train, y_train)

            # Hacer predicciones sobre el conjunto de prueba
            y_pred = log_model.predict(X_test)
            y_pred_prob = log_model.predict_proba(X_test)[:, 1]  # Probabilidades para la curva ROC

            # Calcular y almacenar la precisi√≥n del modelo
            accuracy = metrics.accuracy_score(y_test, y_pred)
            accuracies.append(accuracy)

            # Evaluaci√≥n: Curva ROC y AUC
            fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
            roc_auc = auc(fpr, tpr)
            roc_auc_scores.append(roc_auc)

            # Graficar la curva ROC para cada pliegue
            ax.plot(fpr, tpr, color=colors[fold], lw=2, 
                    label=f'Fold {fold + 1} (AUC = {roc_auc:.2f}')

        # Calcular promedios para el conjunto de hiperpar√°metros actual
        avg_accuracy = np.mean(accuracies)
        avg_roc_auc = np.mean(roc_auc_scores)

        # Guardar resultados
        results.append({
            'penalty': penalty,
            'C': C,
            'avg_accuracy': avg_accuracy,
            'avg_roc_auc': avg_roc_auc
        })

        # Graficar la l√≠nea diagonal (clasificador aleatorio)
        ax.plot([0, 1], [0, 1], linestyle='--', color='gray', lw=2, label="Clasificador aleatorio")

        # Ajustes del subgr√°fico
        ax.set_xlabel('Tasa de Falsos Positivos (FPR)')
        ax.set_ylabel('Tasa de Verdaderos Positivos (TPR)')
        ax.set_title(f'{penalty} (C={C})')
        ax.legend(loc='lower right')
        ax.grid(True, linestyle='--', alpha=0.6)

# Ajustar el layout de la grilla
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Convertir resultados a DataFrame
results_df = pd.DataFrame(results)
```

```python
from sklearn.model_selection import KFold
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm import tqdm  # Importar tqdm para la barra de progreso

# Datos de ejemplo (XFAMD y YFAMD deben estar definidos)
# XFAMD = ...
# YFAMD = ...

# Definir los valores de C, gamma y kernels para la b√∫squeda de hiperpar√°metros
Cs = [0.1, 1.0,]  # Valores de C
gammas = [0.001, 0.01, 0.1]  # Valores de gamma (solo para kernels no lineales)
kernels = ['linear', 'rbf','poly']  # Kernels a probar ,  'sigmoid'

# Configuraci√≥n de K-Fold Cross Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Almacenar resultados
results = []

# Configurar la grilla de gr√°ficos
fig, axes = plt.subplots(len(kernels), len(Cs), figsize=(18, 18))
fig.suptitle('Curvas ROC por Kernel, Valor de C y Gamma - SVM', fontsize=16)

# Definir una paleta de colores para los pliegues
colors = plt.cm.viridis(np.linspace(0, 1, kf.get_n_splits()))

# Contador total de iteraciones
total_iterations = len(kernels) * len(Cs) * kf.get_n_splits()

# Barra de progreso general
with tqdm(total=total_iterations, desc="Progreso General") as pbar:
    # Realizar la b√∫squeda de hiperpar√°metros
    for i, kernel in enumerate(kernels):
        for j, C in enumerate(Cs):
            accuracies = []
            roc_auc_scores = []
            
            # Seleccionar el subgr√°fico actual
            ax = axes[i, j]
            
            for fold, (train_index, test_index) in enumerate(kf.split(XFAMD)):
                # Dividir los datos usando .iloc para acceder a las filas por √≠ndices
                X_train, X_test = XFAMD.iloc[train_index], XFAMD.iloc[test_index]
                y_train, y_test = YFAMD.iloc[train_index], YFAMD.iloc[test_index]

                # Configurar gamma solo para kernels no lineales
                gamma = gammas[j] if kernel in ['poly', 'rbf', 'sigmoid'] else 'scale'

                # Crear el modelo SVM con los hiperpar√°metros actuales
                svm_model = SVC(kernel=kernel, gamma=gamma, C=C, probability=True)
                svm_model.fit(X_train, y_train)

                # Hacer predicciones sobre el conjunto de prueba
                y_pred = svm_model.predict(X_test)
                y_pred_prob = svm_model.predict_proba(X_test)[:, 1]  # Probabilidades para la curva ROC

                # Calcular y almacenar la precisi√≥n del modelo
                accuracy = accuracy_score(y_test, y_pred)
                accuracies.append(accuracy)

                # Evaluaci√≥n: Curva ROC y AUC
                fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
                roc_auc = auc(fpr, tpr)
                roc_auc_scores.append(roc_auc)

                # Graficar la curva ROC para cada pliegue
                ax.plot(fpr, tpr, color=colors[fold], lw=2, 
                        label=f'Fold {fold + 1} AUC = {roc_auc:.2f}')

                # Actualizar la barra de progreso
                pbar.update(1)

            # Calcular promedios para el conjunto de hiperpar√°metros actual
            avg_accuracy = np.mean(accuracies)
            avg_roc_auc = np.mean(roc_auc_scores)

            # Guardar resultados
            results.append({
                'kernel': kernel,
                'C': C,
                'gamma': gamma if kernel in ['poly', 'rbf', 'sigmoid'] else 'N/A',
                'avg_accuracy': avg_accuracy,
                'avg_roc_auc': avg_roc_auc
            })

            # Graficar la l√≠nea diagonal (clasificador aleatorio)
            ax.plot([0, 1], [0, 1], linestyle='--', color='gray', lw=2, label="Clasificador aleatorio")

            # Ajustes del subgr√°fico
            ax.set_xlabel('Tasa de Falsos Positivos (FPR)')
            ax.set_ylabel('Tasa de Verdaderos Positivos (TPR)')
            ax.set_title(f'Kernel={kernel}, C={C}, gamma={gamma}')
            ax.legend(loc='lower right')
            ax.grid(True, linestyle='--', alpha=0.6)

# Ajustar el layout de la grilla
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Convertir resultados a DataFrame
results_df = pd.DataFrame(results)
```
