# Modelos de Clasificación para pacientes con Diabetes

En este análisis, se evaluaron tres modelos diferentes para clasificar a los pacientes diagnosticados con diabetes en función de su probabilidad de ser `readmitidos` al hospital. Los modelos fueron ajustados utilizando validación cruzada con `k-folds` y se compararon mediante gráficos de curvas ROC (Receiver Operating Characteristic) para evaluar su rendimiento. Los modelos utilizados fueron: un modelo GLM binomial, una regresión logística básica y un modelo de máquinas de soporte vectorial (SVM).

## Descripción de los Modelos

1. **Modelo GLM Binomial**:
   - Este modelo utiliza una regresión logística (GLM binomial) para predecir la probabilidad de readmisión de los pacientes. La regresión logística es un método estadístico que modela la relación entre una variable dependiente binaria (en este caso, readmisión o no readmisión) y una o más variables independientes. Es especialmente útil cuando se busca interpretar los coeficientes de las variables predictoras.

2. **Modelo de Regresión Logística Básica**:
   - La regresión logística básica es un enfoque más sencillo y directo para la clasificación binaria. A diferencia del GLM binomial, este modelo no incluye necesariamente una función de enlace generalizada, pero sigue siendo efectivo para predecir la probabilidad de pertenecer a una de las dos clases (readmitido o no readmitido). Es un modelo lineal que funciona bien cuando la relación entre las variables predictoras y la variable objetivo es aproximadamente lineal.

3. **Modelo de Máquinas de Soporte Vectorial (SVM)**:
   - Las máquinas de soporte vectorial (SVM) son un método de aprendizaje supervisado que busca encontrar el hiperplano óptimo que separa las clases en un espacio de características. A diferencia de los modelos lineales, las SVM pueden manejar relaciones no lineales mediante el uso de kernels (por ejemplo, kernel radial o polinómico). Este modelo es especialmente útil cuando los datos no son linealmente separables.

### 1. **Validación Cruzada con K-Folds**

La validación cruzada con \(k\)-folds divide el conjunto de datos en \(k\) subconjuntos (pliegues). En cada iteración:
1. Se entrena el modelo en \(k-1\) pliegues.
2. Se valida en el pliegue restante.
3. Se repite el proceso \(k\) veces, cambiando el pliegue de validación.

Matemáticamente, para cada pliegue \(i\) (\(i = 1, 2, \dots, k\)):
- Conjunto de entrenamiento: \(X_{\text{train}}^{(i)}, y_{\text{train}}^{(i)}\)
- Conjunto de validación: \(X_{\text{test}}^{(i)}, y_{\text{test}}^{(i)}\)

El modelo se ajusta en \(X_{\text{train}}^{(i)}\) y se evalúa en \(X_{\text{test}}^{(i)}\).

### 2. **Métricas de Evaluación**

#### Exactitud (Accuracy)
La exactitud es la proporción de predicciones correctas sobre el total de predicciones:

$$
\text{Accuracy} = \frac{\text{Verdaderos Positivos (TP)} + \text{Verdaderos Negativos (TN)}}{\text{TP} + \text{TN} + \text{Falsos Positivos (FP)} + \text{Falsos Negativos (FN)}}
$$

#### Curva ROC y AUC
La curva ROC (Receiver Operating Characteristic) grafica la **tasa de verdaderos positivos (TPR)** frente a la **tasa de falsos positivos (FPR)** para diferentes umbrales de clasificación:

$$
\text{TPR} = \frac{\text{TP}}{\text{TP} + \text{FN}}, \quad \text{FPR} = \frac{\text{FP}}{\text{FP} + \text{TN}}
$$

El **área bajo la curva ROC (AUC)** es una métrica que resume el rendimiento del modelo:
- AUC = 1: Clasificador perfecto.
- AUC = 0.5: Clasificador aleatorio.

<h3 style="color: #FF7256;">MODELO LINEAL GENERALIZADO BINOMIAL </h3>

### Función de enlace y relación lineal

Un modelo de regresión basado en la distribución binomial (GLM binomial) se define como:

$$
Y_i \sim \text{Bernoulli}(\mu_i)
$$

donde \( \mu_i = E(Y_i | X_i) \) representa la probabilidad condicional del evento de interés. La relación entre \( \mu_i \) y los predictores \( X \) se establece mediante una función de enlace \( g(\cdot) \), de manera que:

$$
g(\mu_i) = \eta_i = X_i \beta
$$

Donde:
- \( g(\cdot) \) es la **función de enlace**, que transforma la media condicional \( \mu_i \) en un predictor lineal \( \eta_i \).
- \( X \) es la **matriz de variables explicativas**.
- \( \beta \) es el **vector de coeficientes** del modelo.
- \( \eta_i \) es el **predictor lineal**.

Para datos agregados donde se modela la cantidad de éxitos en \( n_i \) ensayos, la distribución asumida es:

$$
Y_i \sim \text{Binomial}(n_i, \mu_i)
$$

### Funciones de enlace probadas
En el código, se probaron tres funciones de enlace:

1. **Logit (Enlace logístico)**:
   $$
   g(\mu) = \log\left(\frac{\mu}{1 - \mu}\right)
   $$
   Esta es la función de enlace más común para modelos de regresión logística.

2. **Probit**:
   $$
   g(\mu) = \Phi^{-1}(\mu)
   $$
   Donde \(\Phi^{-1}\) es la función inversa de la distribución acumulada normal estándar.

3. **Cloglog (Complementary log-log)**:
   $$
   g(\mu) = \log(-\log(1 - \mu))
   $$
   Esta función de enlace es útil cuando la probabilidad de éxito es asimétrica.

### Regularización
En el código, se aplicó regularización \(L_2\) (ridge) para evitar el sobreajuste. La función de pérdida regularizada es:

$$
L(\beta) = -\sum_{i=1}^n \left[ y_i \log(\mu_i) + (1 - y_i) \log(1 - \mu_i) \right] + \frac{\alpha}{2} \|\beta\|_2^2
$$

Donde:
- \(y_i\) es la variable dependiente binaria (0 o 1).
- \(\mu_i\) es la probabilidad predicha para la observación \(i\).
- \(\alpha\) es el parámetro de regularización (probaste \(\alpha = 0.001, 0.01, 0.05\)).
- \(\|\beta\|_2^2\) es la norma \(L_2\) del vector de coeficientes.

```python
# ----------------- K-Fold Cross Validation -----------------
kf = KFold(n_splits=5, shuffle=True, random_state=1998)

# Inicialización de variables para almacenar resultados
accuracies = []
roc_auc_scores = []

# Inicializar la figura para las curvas ROC
plt.figure(figsize=(14, 8))

# Definir una paleta de colores
colors = plt.cm.viridis(np.linspace(0, 1, kf.get_n_splits()))

# Realizar K-Fold Cross Validation
for fold, (train_index, test_index) in enumerate(kf.split(XFAMD)):
    # Dividir los datos en conjunto de entrenamiento y prueba
    X_train, X_test = XFAMD.iloc[train_index], XFAMD.iloc[test_index]
    y_train, y_test = YFAMD.iloc[train_index], YFAMD.iloc[test_index]

    # Crear el modelo GLM con familia binomial
    X_train_scaled_sm = sm.add_constant(X_train)
    X_test_scaled_sm = sm.add_constant(X_test)

    # Ajustar el modelo GLM
    glm_binom = sm.GLM(y_train, X_train_scaled_sm, family=sm.families.Binomial())
    res = glm_binom.fit()

    # Realizar predicciones sobre el conjunto de prueba
    y_pred_prob_glm = res.predict(X_test_scaled_sm)
    y_pred_glm = (y_pred_prob_glm > 0.5).astype(int)

    # Evaluación: Precisión
    accuracy = metrics.accuracy_score(y_test, y_pred_glm)
    accuracies.append(accuracy)

    # Evaluación: Curva ROC y AUC
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob_glm)
    roc_auc = auc(fpr, tpr)
    roc_auc_scores.append(roc_auc)

    # Graficar la curva ROC para cada pliegue
    plt.plot(fpr, tpr, color=colors[fold], lw=2, label=f'Fold {fold + 1} (AUC = {roc_auc:.2f})')

avg_accuracy = np.mean(accuracies)
avg_roc_auc = np.mean(roc_auc_scores)


# Graficar la línea diagonal (clasificador aleatorio)
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', lw=2, label="Clasificador aleatorio")

# Ajustes del gráfico
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title(f'Curvas ROC - Validación Cruzada (AUC Promedio = {np.mean(roc_auc_scores):.2f})')
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


print(f"Precisión promedio: {avg_accuracy:.2f}")
print(f"AUC promedio : {avg_roc_auc:.2f}")
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.genmod.families import links
from sklearn.model_selection import KFold
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# Definir las funciones de enlace como objetos
link_functions = {
    'logit': links.Logit(),
    'probit': links.Probit(),
    'cloglog': links.CLogLog()
}

# Valores de alpha para la regularización
alphas = [0.001, 0.01, 0.05]

# Configuración de K-Fold Cross Validation
kf = KFold(n_splits=5, shuffle=True, random_state=1998)

# Almacenar resultados
results = []

# Configurar la grilla de gráficos
fig, axes = plt.subplots(3, 3, figsize=(18, 18))
fig.suptitle('Curvas ROC por Función de Enlace y Alpha', fontsize=16)

# Definir una paleta de colores para los pliegues
colors = plt.cm.viridis(np.linspace(0, 1, kf.get_n_splits()))

# Realizar la búsqueda de hiperparámetros
for i, (link_name, link) in enumerate(link_functions.items()):
    for j, alpha in enumerate(alphas):
        accuracies = []
        roc_auc_scores = []
        
        # Seleccionar el subgráfico actual
        ax = axes[i, j]
        
        for fold, (train_index, test_index) in enumerate(kf.split(XFAMD)):
            # Dividir los datos
            X_train, X_test = XFAMD.iloc[train_index], XFAMD.iloc[test_index]
            y_train, y_test = YFAMD.iloc[train_index], YFAMD.iloc[test_index]

            # Añadir constante
            X_train_scaled_sm = sm.add_constant(X_train)
            X_test_scaled_sm = sm.add_constant(X_test)

            # Crear el modelo GLM con la función de enlace actual
            glm_binom = sm.GLM(y_train, X_train_scaled_sm, 
                               family=sm.families.Binomial(link=link))

            # Ajustar el modelo con regularización
            res = glm_binom.fit_regularized(alpha=alpha)

            # Realizar predicciones
            y_pred_prob_glm = res.predict(X_test_scaled_sm)
            y_pred_glm = (y_pred_prob_glm > 0.5).astype(int)

            # Calcular métricas
            accuracy = metrics.accuracy_score(y_test, y_pred_glm)
            fpr, tpr, _ = roc_curve(y_test, y_pred_prob_glm)
            roc_auc = auc(fpr, tpr)

            # Almacenar resultados del fold
            accuracies.append(accuracy)
            roc_auc_scores.append(roc_auc)

            # Graficar la curva ROC para cada pliegue
            ax.plot(fpr, tpr, color=colors[fold], lw=2, 
                    label=f'Fold {fold + 1} AUC = {roc_auc:.2f}')

        # Calcular promedios para el conjunto de hiperparámetros actual
        avg_accuracy = np.mean(accuracies)
        avg_roc_auc = np.mean(roc_auc_scores)

        # Guardar resultados
        results.append({
            'link': link_name,
            'alpha': alpha,
            'avg_accuracy': avg_accuracy,
            'avg_roc_auc': avg_roc_auc
        })

        # Graficar la línea diagonal (clasificador aleatorio)
        ax.plot([0, 1], [0, 1], linestyle='--', color='gray', lw=2, label="Clasificador aleatorio")

        # Ajustes del subgráfico
        ax.set_xlabel('Tasa de Falsos Positivos (FPR)')
        ax.set_ylabel('Tasa de Verdaderos Positivos (TPR)')
        ax.set_title(f'{link_name} (alpha={alpha})')
        ax.legend(loc='lower right')
        ax.grid(True, linestyle='--', alpha=0.6)

# Ajustar el layout de la grilla
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Convertir resultados a DataFrame
results_df = pd.DataFrame(results)
```

<h2 style="color: black;">REGRESION LOGISTICA</h2>

El modelo de regresión logística es un método de clasificación binaria que modela la probabilidad de pertenecer a una clase (en este caso, la readmisión de pacientes). La estructura matemática del modelo es la siguiente:

### Función de predicción
La probabilidad de que la variable dependiente $Y$ sea 1 (readmitido) dado un conjunto de predictores $X$ se modela como:

$$
P(Y = 1 | X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_p X_p)}}
$$

Donde:
- $\beta_0$ es el término de intercepción.
- $\beta_1, \beta_2, \dots, \beta_p$ son los coeficientes asociados a las variables predictoras $X_1, X_2, \dots, X_p$.
- $e$ es la base del logaritmo natural.

### Función de pérdida
La función de pérdida para la regresión logística es la **entropía cruzada binaria** (binary cross-entropy):

$$
L(\beta) = -\sum_{i=1}^n \left[ y_i \log(p_i) + (1 - y_i) \log(1 - p_i) \right]
$$

Donde:
- $y_i$ es la etiqueta real (0 o 1) para la observación $i$.
- $p_i = P(Y = 1 | X_i)$ es la probabilidad predicha para la observación $i$.

### Regularización
En el código, se aplicaron dos tipos de regularización:

1. **Regularización $L_1$ (Lasso)**:
   $$
   L(\beta) = -\sum_{i=1}^n \left[ y_i \log(p_i) + (1 - y_i) \log(1 - p_i) \right] + \lambda \|\beta\|_1
   $$
   Donde \(\|\beta\|_1\) es la norma \(L_1\) del vector de coeficientes.

2. **Regularización $L_2$ (Ridge)**:
   $$
   L(\beta) = -\sum_{i=1}^n \left[ y_i \log(p_i) + (1 - y_i) \log(1 - p_i) \right] + \lambda \|\beta\|_2^2
   $$
   Donde $\|\beta\|_2^2$ es la norma $L_2$ del vector de coeficientes.



En el código, $\lambda = \frac{1}{C}$, donde $C$ es el parámetro de regularización inversa (se probo $C = 0.01, 0.05, 0.1$).

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Definir los valores de C (inverso de la fuerza de regularización) y los tipos de regularización
Cs = [0.01, 0.05, 0.1]  # Valores de C
penalties = ['l1', 'l2']  # Tipos de regularización

# Configuración de K-Fold Cross Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Almacenar resultados
results = []

# Configurar la grilla de gráficos
fig, axes = plt.subplots(len(penalties), len(Cs), figsize=(18, 12))
fig.suptitle('Curvas ROC por Tipo de Regularización y Valor de C', fontsize=16)

# Definir una paleta de colores para los pliegues
colors = plt.cm.viridis(np.linspace(0, 1, kf.get_n_splits()))

# Realizar la búsqueda de hiperparámetros
for i, penalty in enumerate(penalties):
    for j, C in enumerate(Cs):
        accuracies = []
        roc_auc_scores = []
        
        # Seleccionar el subgráfico actual
        ax = axes[i, j]
        
        for fold, (train_index, test_index) in enumerate(kf.split(XDAT_encoded)):
            # Dividir los datos usando .iloc para acceder a las filas por índices
            X_train, X_test = XDAT_encoded.iloc[train_index], XDAT_encoded.iloc[test_index]
            y_train, y_test = YDAT.iloc[train_index], YDAT.iloc[test_index]

            # Crear el modelo de regresión logística
            log_model = LogisticRegression(penalty=penalty, C=C, solver='liblinear')
            log_model.fit(X_train, y_train)

            # Hacer predicciones sobre el conjunto de prueba
            y_pred = log_model.predict(X_test)
            y_pred_prob = log_model.predict_proba(X_test)[:, 1]  # Probabilidades para la curva ROC

            # Calcular y almacenar la precisión del modelo
            accuracy = metrics.accuracy_score(y_test, y_pred)
            accuracies.append(accuracy)

            # Evaluación: Curva ROC y AUC
            fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
            roc_auc = auc(fpr, tpr)
            roc_auc_scores.append(roc_auc)

            # Graficar la curva ROC para cada pliegue
            ax.plot(fpr, tpr, color=colors[fold], lw=2, 
                    label=f'Fold {fold + 1} (AUC = {roc_auc:.2f}')

        # Calcular promedios para el conjunto de hiperparámetros actual
        avg_accuracy = np.mean(accuracies)
        avg_roc_auc = np.mean(roc_auc_scores)

        # Guardar resultados
        results.append({
            'penalty': penalty,
            'C': C,
            'avg_accuracy': avg_accuracy,
            'avg_roc_auc': avg_roc_auc
        })

        # Graficar la línea diagonal (clasificador aleatorio)
        ax.plot([0, 1], [0, 1], linestyle='--', color='gray', lw=2, label="Clasificador aleatorio")

        # Ajustes del subgráfico
        ax.set_xlabel('Tasa de Falsos Positivos (FPR)')
        ax.set_ylabel('Tasa de Verdaderos Positivos (TPR)')
        ax.set_title(f'{penalty} (C={C})')
        ax.legend(loc='lower right')
        ax.grid(True, linestyle='--', alpha=0.6)

# Ajustar el layout de la grilla
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Convertir resultados a DataFrame
results_df = pd.DataFrame(results)
```

<h3 style="color: black;">MAQUINA DE SOPORTE VECTORIAL</h3>

El modelo SVM es un método de aprendizaje supervisado utilizado para clasificación y regresión. En este caso, se utiliza para clasificación binaria (readmisión de pacientes). La estructura matemática del modelo depende del kernel utilizado.

### Función de decisión
La función de decisión de un SVM para un punto $x$ es:

$$
f(x) = \text{sign}\left(\sum_{i=1}^n \alpha_i y_i K(x_i, x) + b\right)
$$

Donde:
- $\alpha_i$ son los coeficientes de Lagrange (obtenidos durante el entrenamiento).
- $y_i$ son las etiquetas de clase ($-1$ o $1$).
- $K(x_i, x)$ es la función kernel que mide la similitud entre $x_i$ y $x$.
- $b$ es el término de sesgo.

### Función de pérdida
La función de pérdida para SVM es la **pérdida de hinge**:

$$
L(\mathbf{w}, b) = \frac{1}{2} \|\mathbf{w}\|^2 + C \sum_{i=1}^n \max(0, 1 - y_i (\mathbf{w} \cdot \mathbf{x}_i + b))
$$

Donde:
- $\mathbf{w}$ es el vector de pesos.
- $C$ es el parámetro de regularización (inverso de la fuerza de regularización).
- $y_i$ es la etiqueta real ($-1$ o $1$).
- $\mathbf{x}_i$ es el vector de características.

### Kernels probados
En el código, se probaron tres tipos de kernels:

1. **Kernel Lineal**:
   $$
   K(x_i, x_j) = x_i \cdot x_j
   $$
   Este kernel es adecuado cuando los datos son linealmente separables.

2. **Kernel RBF (Radial Basis Function)**:
   $$
   K(x_i, x_j) = \exp\left(-\gamma \|x_i - x_j\|^2\right)
   $$
   Donde $\gamma$ es un parámetro que controla el alcance de la influencia de cada punto.

3. **Kernel Polinómico**:
   $$
   K(x_i, x_j) = (\gamma (x_i \cdot x_j) + r)^d
   $$
   Donde $d$ es el grado del polinomio y $r$ es un término independiente.

### Regularización
El parámetro $C$ controla el equilibrio entre la maximización del margen y la minimización del error de clasificación:
- $C$ pequeño: Mayor regularización, margen más amplio.
- $C$ grande: Menor regularización, margen más estrecho.

En el código, se probaron valores de $C = [0.1, 1.0]$ y $\gamma = [0.001, 0.01, 0.1]$.

```python
from sklearn.model_selection import KFold
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm import tqdm 


# Definir los valores de C, gamma y kernels para la búsqueda de hiperparámetros
Cs = [0.1, 1.0,]  # Valores de C
gammas = [0.001, 0.01, 0.1]  # Valores de gamma (solo para kernels no lineales)
kernels = ['linear', 'rbf','poly']  # Kernels a probar ,  'sigmoid'

# Configuración de K-Fold Cross Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Almacenar resultados
results = []

# Configurar la grilla de gráficos
fig, axes = plt.subplots(len(kernels), len(Cs), figsize=(18, 18))
fig.suptitle('Curvas ROC por Kernel, Valor de C y Gamma - SVM', fontsize=16)

# Definir una paleta de colores para los pliegues
colors = plt.cm.viridis(np.linspace(0, 1, kf.get_n_splits()))

# Contador total de iteraciones
total_iterations = len(kernels) * len(Cs) * kf.get_n_splits()

# Barra de progreso general
with tqdm(total=total_iterations, desc="Progreso General") as pbar:
    # Realizar la búsqueda de hiperparámetros
    for i, kernel in enumerate(kernels):
        for j, C in enumerate(Cs):
            accuracies = []
            roc_auc_scores = []
            
            # Seleccionar el subgráfico actual
            ax = axes[i, j]
            
            for fold, (train_index, test_index) in enumerate(kf.split(XFAMD)):
                # Dividir los datos usando .iloc para acceder a las filas por índices
                X_train, X_test = XFAMD.iloc[train_index], XFAMD.iloc[test_index]
                y_train, y_test = YFAMD.iloc[train_index], YFAMD.iloc[test_index]

                # Configurar gamma solo para kernels no lineales
                gamma = gammas[j] if kernel in ['poly', 'rbf', 'sigmoid'] else 'scale'

                # Crear el modelo SVM con los hiperparámetros actuales
                svm_model = SVC(kernel=kernel, gamma=gamma, C=C, probability=True)
                svm_model.fit(X_train, y_train)

                # Hacer predicciones sobre el conjunto de prueba
                y_pred = svm_model.predict(X_test)
                y_pred_prob = svm_model.predict_proba(X_test)[:, 1]  # Probabilidades para la curva ROC

                # Calcular y almacenar la precisión del modelo
                accuracy = accuracy_score(y_test, y_pred)
                accuracies.append(accuracy)

                # Evaluación: Curva ROC y AUC
                fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
                roc_auc = auc(fpr, tpr)
                roc_auc_scores.append(roc_auc)

                # Graficar la curva ROC para cada pliegue
                ax.plot(fpr, tpr, color=colors[fold], lw=2, 
                        label=f'Fold {fold + 1} AUC = {roc_auc:.2f}')

                # Actualizar la barra de progreso
                pbar.update(1)

            # Calcular promedios para el conjunto de hiperparámetros actual
            avg_accuracy = np.mean(accuracies)
            avg_roc_auc = np.mean(roc_auc_scores)

            # Guardar resultados
            results.append({
                'kernel': kernel,
                'C': C,
                'gamma': gamma if kernel in ['poly', 'rbf', 'sigmoid'] else 'N/A',
                'avg_accuracy': avg_accuracy,
                'avg_roc_auc': avg_roc_auc
            })

            # Graficar la línea diagonal (clasificador aleatorio)
            ax.plot([0, 1], [0, 1], linestyle='--', color='gray', lw=2, label="Clasificador aleatorio")

            # Ajustes del subgráfico
            ax.set_xlabel('Tasa de Falsos Positivos (FPR)')
            ax.set_ylabel('Tasa de Verdaderos Positivos (TPR)')
            ax.set_title(f'Kernel={kernel}, C={C}, gamma={gamma}')
            ax.legend(loc='lower right')
            ax.grid(True, linestyle='--', alpha=0.6)

# Ajustar el layout de la grilla
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Convertir resultados a DataFrame
results_df = pd.DataFrame(results)
```
